# Part 1

```python
with open("day13_input.txt") as f:
    lines = f.read().splitlines()
    
groups_of_mirror = []

group_mirror  = []
for i, line in enumerate(lines):
    if line != "":
        group_mirror.append(line)
    if line == "" or i+1 == len(lines):
        groups_of_mirror.append(group_mirror)
        group_mirror =[]

for i, mirror_set in enumerate(groups_of_mirror):
    groups_of_mirror[i] = np.array([list(line) for line in mirror_set])

def find_horizontal(mirror_set):
    mirror_lines =[]
    for i, row in enumerate(mirror_set):
        #print(i, row)
        if i+1 < len(mirror_set):
            if np.array_equal(mirror_set[i], mirror_set[i+1]):
                #print("Mirror line found", i )
                mirror_line = i 
                mirror_lines.append(mirror_line)

    if mirror_lines==[]:
        return 0

    for x, mirror_line in enumerate(mirror_lines):
        #print("Checking mirror_line", mirror_line)
        if mirror_line == 0:
            return 1
        elif mirror_line==mirror_set.shape[0]-2:
            return mirror_set.shape[0]-1
        
        for i in range(0, mirror_line):
            distance_from_line = mirror_line-i
            current_index = i 
            mirror_index = mirror_line+distance_from_line+1
            if mirror_index < mirror_set.shape[0]:
                #print(current_index, mirror_index, mirror_set[current_index], mirror_set[mirror_index])
                if not np.array_equal(mirror_set[current_index], mirror_set[mirror_index]):
                    #print("No perfect reflection for mirror line", mirror_line)
                    if x == len(mirror_lines)-1:
                        return 0
                    else:
                        break
                elif i == mirror_line-1:
                    #print("Perfectly matching")
                    return mirror_line+1
        
def find_vertical(mirror_set):
    mirror_lines =[]

    for i in range(0, mirror_set.shape[1]):
        #print(i, mirror_set[:,i])
        if i+1 < mirror_set.shape[1]:
            if np.array_equal(mirror_set[:,i], mirror_set[:,i+1]):
                #print("Mirror line found", i )
                mirror_line = i 
                mirror_lines.append(mirror_line)

    if mirror_lines==[]:
        return 0

    for x, mirror_line in enumerate(mirror_lines):
        #print("Checking mirror_line", mirror_line)
        if mirror_line == 0:
            return 1
        elif mirror_line==mirror_set.shape[1]-2:
            return mirror_set.shape[1]-1
        
        for i in range(0, mirror_line):         
            distance_from_line = mirror_line-i
            current_index = i 
            mirror_index = mirror_line+distance_from_line+1
            #print("i", i, "mirror index", mirror_index)
            if mirror_index < mirror_set.shape[1]:
                if not np.array_equal(mirror_set[:,current_index], mirror_set[:,mirror_index]):
                    #print("No perfect reflection")
                    if x == len(mirror_lines)-1:
                        return 0
                    else:
                        break
                elif i == mirror_line-1:
                    #print("Perfectly matching")
                    return mirror_line+1

horizontal_sum = 0
vertical_sum = 0

for i, mirror_set in enumerate(groups_of_mirror):
    #print("Checking ", i)
    #display(mirror_set)
    #print("H", find_horizontal(mirror_set), "V", find_vertical(mirror_set))
    horizontal_sum+=find_horizontal(mirror_set)
    vertical_sum+=find_vertical(mirror_set)

print(horizontal_sum*100+vertical_sum)
```

# Part 2

```python
with open("day13_input.txt") as f:
    lines = f.read().splitlines()
    
groups_of_mirror = []

group_mirror  = []
for i, line in enumerate(lines):
    if line != "":
        group_mirror.append(line)
    if line == "" or i+1 == len(lines):
        groups_of_mirror.append(group_mirror)
        group_mirror =[]

for i, mirror_set in enumerate(groups_of_mirror):
    groups_of_mirror[i] = np.array([list(line) for line in mirror_set])

def swap_char(char_mirror):
    if char_mirror == "#":
        return "."
    else:
        return "#"
    
def find_horizontal_mirror_lines(mirror_set):
    mirror_lines =[]
    for i, row in enumerate(mirror_set):
        #print(i, row)
        if i+1 < len(mirror_set):
            if np.array_equal(mirror_set[i], mirror_set[i+1]):
                #print("Mirror line found", i )
                mirror_line = i 
                mirror_lines.append(mirror_line)
    return mirror_lines

def find_horizontal_new(mirror_set, mirror_lines):
    if mirror_lines==[]:
        return 0

    for x, mirror_line in enumerate(mirror_lines):
        #print("Checking mirror_line", mirror_line)
        if mirror_line == 0:
            return 1
        elif mirror_line==mirror_set.shape[0]-2:
            return mirror_set.shape[0]-1
        
        for i in range(0, mirror_line):
            distance_from_line = mirror_line-i
            current_index = i 
            mirror_index = mirror_line+distance_from_line+1
            if mirror_index < mirror_set.shape[0]:
                #print(current_index, mirror_index, mirror_set[current_index], mirror_set[mirror_index])
                if not np.array_equal(mirror_set[current_index], mirror_set[mirror_index]):
                    #print("No perfect reflection for mirror line", mirror_line)
                    if x == len(mirror_lines)-1:
                        return 0
                    else:
                        break
                elif i == mirror_line-1:
                    #print("Perfectly matching")
                    return mirror_line+1

def find_vertical_mirror_lines(mirror_set):
    mirror_lines =[]
    for i in range(0, mirror_set.shape[1]):
        #print(i, mirror_set[:,i])
        if i+1 < mirror_set.shape[1]:
            if np.array_equal(mirror_set[:,i], mirror_set[:,i+1]):
                #print("Mirror line found", i )
                mirror_line = i 
                mirror_lines.append(mirror_line)
    return mirror_lines

def find_vertical_new(mirror_set, mirror_lines):
    if mirror_lines==[]:
        return 0
    
    #print("Mirror lines", mirror_lines)
    for x, mirror_line in enumerate(mirror_lines):
        #print("Checking mirror_line", mirror_line)
        if mirror_line == 0:
            return 1
        elif mirror_line==mirror_set.shape[1]-2:
            return mirror_set.shape[1]-1
        
        for i in range(0, mirror_line):         
            distance_from_line = mirror_line-i
            current_index = i 
            mirror_index = mirror_line+distance_from_line+1
            #print("i", i, "mirror index", mirror_index)
            if mirror_index < mirror_set.shape[1]:
                if not np.array_equal(mirror_set[:,current_index], mirror_set[:,mirror_index]):
                    #print("No perfect reflection")
                    if x == len(mirror_lines)-1:
                        return 0
                    else:
                        break
                elif i == mirror_line-1:
                    #print("Perfectly matching")
                    return mirror_line+1
                
def check_smudge(mirror_set):
    hori_init, vertical_init = find_horizontal(mirror_set), find_vertical(mirror_set)

    for x in range(0, mirror_set.shape[0]):
        for y in range(0, mirror_set.shape[1]):
            mirror_set_new = mirror_set.copy()
            mirror_set_new[x,y] = swap_char(mirror_set_new[x,y])

            horizontal_mirror_lines = list(set(find_horizontal_mirror_lines(mirror_set_new))-set([find_horizontal(mirror_set)-1]))
            vertical_mirror_lines = list(set(find_vertical_mirror_lines(mirror_set_new))-set([find_vertical(mirror_set)-1]))

            hori_new, vertical_new = find_horizontal_new(mirror_set_new, horizontal_mirror_lines), find_vertical_new(mirror_set_new, vertical_mirror_lines)
            #print(x, y, "..", hori_new, vertical_new)
            if hori_new!=0 and hori_new!=hori_init:
                #print("different horizontal found changing", x, y)
                #display(mirror_set_new)
                return "hori", hori_new
            if vertical_new!=0 and vertical_new!=vertical_init:
                #print("different vertical found changing", x, y)
                #display(mirror_set_new)
                return "vert", vertical_new            

horizontal_sum = 0
vertical_sum = 0
for i, mirror_set in enumerate(groups_of_mirror):
    type_refl, change =check_smudge(mirror_set)
    #print("i", i, "change", change)

    if type_refl == "hori":
        horizontal_sum +=change
    if type_refl == "vert":
        vertical_sum += change

print(horizontal_sum*100+vertical_sum)
```
