# Part 1

```python
with open("day1_input.txt") as f:
    instructions = f.read().splitlines() 
```

```python
starting_positions = 50

def rotate_dial(curent_position, instruction):
    direction = instruction[0]
    steps = int(instruction[1:])


    if direction == "R":
        new_position = (curent_position + steps)%100
    else:
        new_position = (curent_position - steps)%100

    # print(f"Current position: {curent_position}, Direction: {direction}, Steps: {steps}, new position: {new_position}")
    return new_position

current_position = starting_positions

counter_zeros = 0
for instruction in instructions:
    current_position = rotate_dial(current_position, instruction)
    if current_position == 0:
        counter_zeros += 1

print(f"Pt1: Final position: {current_position}, Number of times at 0: {counter_zeros}")
```

# Part 2

```python
starting_positions = 50

def rotate_dial(curent_position, instruction):
    direction = instruction[0]
    steps = int(instruction[1:])


    if direction == "R":
        new_position = (curent_position + steps)%100
    else:
        new_position = (curent_position - steps)%100

    print(f"Current position: {curent_position}, Direction: {direction}, Steps: {steps}, new position: {new_position}")
    return new_position

current_position = starting_positions

counter_zeros = 0
for instruction in instructions:
    direction = instruction[0]
    steps = int(instruction[1:])
    starting_position = current_position
    current_position = rotate_dial(current_position, instruction)

    if current_position == 0:
        counter_zeros += 1
        print("Curent position is zero")
    elif starting_position != 0:
        if direction == "R":
            sign = 1
        elif direction == "L":
            sign = -1

        if abs((starting_position+(steps*sign))//100)>0:
            print(f"Crossed zero between {starting_position} and {current_position}")
            print(abs((starting_position+(steps*sign))//100))
            counter_zeros += 1  

print(f"Pt2: Final position: {current_position}, Number of times at 0: {counter_zeros}")
```
