```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', True)

import re
```

```python
with open("day2_input.txt") as f:
    lines = f.read().splitlines() 

reports = []
for line in [x.split() for x in lines]:
    reports.append([int(x) for x in line])
```

# Part 1

```python
def check_levels(report):
    differences = [report[i+1]-report[i]  for i in range(len(report)-1)]
    print(differences)
    if abs(sum([np.sign(x) for x in differences]))!=len(differences):
        print("not safe - incorrect sign")
        return False
    else:
        if max(abs(np.array(differences)))<=3 and min(abs(np.array(differences)))>=1:
            print("safe")
            return True
        else:
            print("not safe - max or min number too high or too low")
            return False


total_safe = 0
for i, report in enumerate(reports):
    print(i, report)
    if check_levels(report)==True:
        total_safe+=1

print(total_safe)


```

# Part 2

```python
def check_levels(report):
    differences = [report[i+1]-report[i]  for i in range(len(report)-1)]
    print("differences", differences)
    if abs(sum([np.sign(x) for x in differences]))!=len(differences):
        print("not safe - incorrect sign")
        return False
    else:
        if max(abs(np.array(differences)))<=3 and min(abs(np.array(differences)))>=1:
            print("safe")
            return True
        else:
            print("not safe - max or min number too high or too low")
            return False


total_safe = 0
for i, report in enumerate(reports):
    print("---------", i, report)
    if check_levels(report)==True:
        total_safe+=1
    else:
        # try to remove one level at the time:
        for x in range(0,len(report)):
            short_report = report[0:x]+report[x+1:]
            print("Checking possibility ", short_report)
            if check_levels(short_report)==True:
                total_safe+=1
                print("new report is safe ", short_report)
                break

print(total_safe)
```
