```python
import pandas as pd
import numpy as np
import math
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', True)
from tqdm import tqdm
from itertools import product
from itertools import combinations

import re
```

```python
def read_matrix_from_file(file_path):
    with open(file_path, 'r') as file:
        input_text = file.read()
        lines = input_text.strip().split('\n')
        matrix = [list(line) for line in lines]
        numpy_matrix = np.array(matrix)
        return numpy_matrix

map = read_matrix_from_file("day8_input.txt") 
```

# Part1

```python
x_size = len(map)
y_size = len(map[0])
print("Size map", x_size, y_size)

def find_antinodes(coord1,coord2):
    x1,y1,x2,y2= coord1[0], coord1[1],coord2[0], coord2[1]
    dx = x2-x1
    dy = y2-y1
    x2_an, y2_an = x2+dx, y2+dy
    x1_an, y1_an = x1-dx, y1-dy

    #print("Antinode1", x1_an, y1_an, "Antinode2",x2_an, y2_an)
    valid_antinodes = []
    if x1_an>=0 and x1_an<x_size and y1_an>=0 and y1_an<y_size:
        valid_antinodes.append((x1_an,y1_an))
    if x2_an>=0 and x2_an<x_size and y2_an>=0 and y2_an<y_size:
        valid_antinodes.append((x2_an,y2_an))
    return valid_antinodes

unique_anthennas = set(np.unique(map))-set(".")
print("unique_anthennas", unique_anthennas)

unique_antinodes = []
for anthenna_type in unique_anthennas:
    print("Checking anthenna type", anthenna_type)
    positions = np.where(map == anthenna_type)
    # Output as pairs of coordinates
    coordinates = list(zip(positions[0], positions[1]))
    print(coordinates)

    pairs_anthennas = list(combinations(coordinates, 2))
    #print(f"Pair of anthennas for {anthenna_type}, {pairs_anthennas}")
    # Iterate through the pairs

    for pair in pairs_anthennas:
        #print("Checking", pair)
        #print("valid antinodes",find_antinodes(pair[0], pair[1]))
        unique_antinodes +=find_antinodes(pair[0], pair[1])

print("Signal impact", len(set(unique_antinodes)))
print(set(unique_antinodes))
```

# Part 2

```python
def find_antinodes_new(coord1,coord2, map):
    x_size = len(map)
    y_size = len(map[0])
    x1,y1,x2,y2= coord1[0], coord1[1],coord2[0], coord2[1]
    #print("checking ", x1,y1,x2,y2)
    dx = x2-x1
    dy = y2-y1
    dx_max1, dy_max1 = x_size-x1, y_size-y1
    dx_max2, dy_max2 = x_size-x1, y_size-y1

    valid_antinodes = []
    iter = 1
    additional_antinodes1 = True
    additional_antinodes2 = True

    # Keep iterating until there are no more valid antinodes to include
    while additional_antinodes1 or additional_antinodes2:
   
        x2_an, y2_an = x2+dx*iter, y2+dy*iter
        x1_an, y1_an = x1-dx*iter, y1-dy*iter
        #print("Antinode1", x1_an, y1_an, "Antinode2",x2_an, y2_an)
        
        if x1_an>=0 and x1_an<x_size and y1_an>=0 and y1_an<y_size:
            valid_antinodes.append((x1_an,y1_an))
            additional_antinodes1 = True
        else:
            #print("antinode 1 not valid")
            additional_antinodes1 = False

        if x2_an>=0 and x2_an<x_size and y2_an>=0 and y2_an<y_size:
            valid_antinodes.append((x2_an,y2_an))
            additional_antinodes2 = True
        else:
            additional_antinodes2 = False
            #print("antinode 2 not valid")
            
        iter+=1
    
    return valid_antinodes


unique_anthennas = set(np.unique(map))-set(".")
print("unique_anthennas", unique_anthennas)

unique_antinodes = []
for anthenna_type in unique_anthennas:
    print("Checking anthenna type", anthenna_type)
    positions = np.where(map == anthenna_type)
    # Output as pairs of coordinates
    coordinates = list(zip(positions[0], positions[1]))
    print(coordinates)

    pairs_anthennas = list(combinations(coordinates, 2))
    #print(f"Pair of anthennas for {anthenna_type}, {pairs_anthennas}")
    # Iterate through the pairs

    for pair in pairs_anthennas:
        #print("Checking", pair)
        #print("valid antinodes",find_antinodes(pair[0], pair[1]))
        unique_antinodes +=find_antinodes_new(pair[0], pair[1], map)
        unique_antinodes += pair

print("Signal impact", len(set(unique_antinodes)))
print(set(unique_antinodes))
```
