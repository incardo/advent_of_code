```python
import pandas as pd
import numpy as np
import math
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', True)
from tqdm import tqdm
from itertools import product
import re
```

```python
with open("day7_input.txt") as f:
    lines = f.read().split('\n')

equations = []
for i, line in enumerate(lines):
    result = int(line.split(":")[0])
    numbers = [int(x) for x in line.split(":")[1].strip().split(" ")]
    #print(result, numbers)
    equations.append([result, numbers])

```

# Part 1

```python
operators_general = ['*', '+']
correct_total_calibration_result = 0
for nr_eq, equation in enumerate(equations):
    result = equation[0]
    numbers = equation[1]

    #print("Checking", result, numbers)

    # Generate all permutations of operators (duplicates allowed)
    permutations = list(product(operators_general, repeat=len(numbers)-1))
    permutations_operators = [','.join(p) for p in permutations]
    #print(permutations_operators)

    count_correct = 0
    for operators in permutations_operators:
        operators = operators.split(",")

        # Sequential evaluation
        operation_to_eval = numbers[0]
        for i in range(len(operators)):
            if operators[i]=="+":
                operation_to_eval+= numbers[i+1]
            elif operators[i]=="*":
                operation_to_eval*= numbers[i+1]
        #print("result checking", operation_to_eval, operators)
        if  operation_to_eval==result:
            #print(operators , "correct")
            count_correct+=1
    
    if count_correct>0:
        #print("Equation is fine", count_correct, result, )
        correct_total_calibration_result += result
    else:
        #print("Equation not fine", count_correct)
        continue

print(correct_total_calibration_result)

    
```

# Part 2

```python
operators_general = ['*', '+', "||"]
correct_total_calibration_result = 0
for nr_eq, equation in enumerate(equations):
    result = equation[0]
    numbers = equation[1]

    print("Checking", nr_eq, result, numbers)

    # Generate all permutations of operators (duplicates allowed)
    permutations = list(product(operators_general, repeat=len(numbers)-1))
    permutations_operators = [','.join(p) for p in permutations]
    #print(permutations_operators)

    count_correct = 0
    for operators in permutations_operators:
        operators = operators.split(",")

        # Sequential evaluation
        operation_to_eval = numbers[0]
        for i in range(len(operators)):
            if operators[i]=="+":
                operation_to_eval+= numbers[i+1]
            elif operators[i]=="*":
                operation_to_eval*= numbers[i+1]
            elif operators[i] == "||":
                operation_to_eval = int(str(operation_to_eval)+str(numbers[i+1]))

        #print("result checking", operation_to_eval, operators)
        if  operation_to_eval==result:
            #print(operators , "correct")
            count_correct+=1
    
    if count_correct>0:
        #print("Equation is fine", count_correct, result, )
        correct_total_calibration_result += result
    else:
        #print("Equation not fine", count_correct)
        continue

print(correct_total_calibration_result)

    
```
