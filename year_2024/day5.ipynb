```python
import pandas as pd
import numpy as np
import math
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', True)

import re
```

```python
with open("day5_input.txt") as f:
    lines = f.read().split('\n')

rules = [x for x in lines if "|" in x]
updates = [x for x in lines if "," in x]
```

# Part 1

```python
valid_updates = []
invalid_updates = []
for update in updates:
    nrs_update = update.split(",")
    #print(nrs_update)

    errors = False
    for i, nr in enumerate(nrs_update):
        #print(i, nr)
        if i+1!=len(nrs_update):
            theoretical_rule = nr+"|"+nrs_update[i+1]
            if theoretical_rule in rules:
                #print(f"{nr,nrs_update[i+1]} correct")
                pass
            else:
                #print(f"Error with {update}")
                errors = True
                break
    if not errors:
        valid_updates.append(update)
    else:
        invalid_updates.append(update)

valid_updates

middle_page_sum = 0
for update in valid_updates:
    update = update.split(",")
    #print(update,  int(update[math.floor((len(update)/2))]))
    middle_page_sum+= int(update[math.floor((len(update)/2))])

middle_page_sum
```

```python
fixed_updates = []
for update in invalid_updates:
    nrs_update = update.split(",")
    #print("START", nrs_update)

    reordered_update = nrs_update.copy()
    incorrect_order = True

    # Keep resorting until all numbers are in the right order
    while incorrect_order:
        for i, nr in enumerate(reordered_update):
            #print(i, nr)
            if i+1!=len(reordered_update):
                theoretical_rule = nr+"|"+reordered_update[i+1]
                #print("theoretical_rule", theoretical_rule)
                if theoretical_rule in rules:
                    #print(f"{nr,nrs_update[i+1]} correct")
                    incorrect_order = False
                    pass
                else:
                    # Try to fix order of numbers
                    reordered_update[i]=reordered_update[i+1]
                    reordered_update[i+1]=nr
                    incorrect_order = True
                    #print("There was a mistake: New order", reordered_update, "restarting...")
                    break
    
    #print("Are still incorrect orders?", incorrect_order)
    
    fixed_updates.append(reordered_update)

#display("corrected", fixed_updates)           
middle_page_sum = 0
for update in fixed_updates:
    #print(update,  int(update[math.floor((len(update)/2))]))
    middle_page_sum+= int(update[math.floor((len(update)/2))])

middle_page_sum
```
